# Glosario de Conceptos

### A - Z



BOOLEANOS (bool)
Tipo de dato lógico que puede tener únicamente dos valores: True (Verdadero) o False (Falso).

CICLOS (Loops)
Estructuras de control que repiten la ejecución de un bloque de código.
  - for: Itera sobre una secuencia predefinida (como una lista o string).
  - while: Repite un bloque de código mientras una condición sea True.
  - break: Interrumpe y finaliza el ciclo por completo.
  - continue: Salta el resto del código en la iteración actual y pasa a la siguiente.
  - range(): Genera una secuencia de números enteros (ej. range(5) genera del 0 al 4).
  - enumerate(): Itera sobre una secuencia devolviendo tanto el índice como el valor en cada paso.

CONDICIONALES
Estructuras lógicas (if, elif, else) que permiten que el flujo del código tome distintas ramas basándose en la evaluación de una condición.

CONJUNTOS (Sets)
Colección de datos que almacena elementos sin ningún orden en particular y no permite elementos duplicados.

DATAFRAME
Estructura de datos bidimensional provista por Pandas, similar a una tabla, fundamental para el análisis de datos.

DICCIONARIOS (dict)
Estructura de almacenamiento que utiliza una correspondencia de pares "clave: valor" en lugar de índices numéricos.

ENTEROS (int)
Tipo de variable numérica que representa números exactos sin parte decimal.

ERRORES COMUNES
Excepciones que interrumpen el código cuando algo falla:
  - NameError: Ocurre cuando se intenta usar una variable o función que no ha sido definida.
  - TypeError: Ocurre al aplicar una operación a un tipo de dato inadecuado (ej. sumar texto y números).
  - IndexError: Ocurre cuando se intenta acceder a un índice que no existe en una lista o tupla.
  - KeyError: Ocurre al buscar una clave que no existe en un diccionario.
  - ModuleNotFoundError: Ocurre cuando se intenta importar una librería que no está instalada o está mal escrita.

FLOTANTES (float)
Tipo de variable numérica que representa números que incluyen una parte decimal.

FUNCIONES
Bloque de código empaquetado y reutilizable diseñado para realizar una tarea específica.
  - def: Palabra reservada que se utiliza para definir o crear una función.
  - return: Palabra reservada que especifica el valor o resultado que la función debe devolver al terminar.
  - parámetros: Variables de entrada que recibe la función para trabajar con ellas internamente.

LISTAS (list)
Estructura de datos mutable que almacena una colección ordenada de elementos. Se declaran con [].
  - append(): Agrega un elemento al final de la lista.
  - pop(): Elimina y devuelve un elemento (por defecto el último, o el del índice indicado).
  - remove(): Elimina la primera aparición de un valor específico en la lista.
  - insert(): Inserta un elemento en una posición (índice) específica.
  - extend(): Agrega múltiples elementos (iterables) al final de la lista.
  - sort(): Ordena los elementos de la lista (alfabética o numéricamente).
  - reverse(): Invierte el orden actual de los elementos en la lista.

MATPLOTLIB
Biblioteca de Python utilizada para la visualización de datos.
  - plt.figure(): Crea una nueva figura o espacio de trabajo para el gráfico.
  - plt.plot(): Crea un gráfico de líneas.
  - plt.hist(): Crea un histograma (distribución de frecuencias).
  - plt.bar(): Crea un gráfico de barras.
  - plt.xlabel() / plt.ylabel(): Asignan etiquetas de texto al eje X y al eje Y respectivamente.
  - plt.title(): Agrega un título principal al gráfico.
  - plt.grid(): Muestra una cuadrícula de fondo en el gráfico para facilitar la lectura.
  - plt.show(): Renderiza y muestra el gráfico en pantalla.

NUMPY
Biblioteca especializada en cálculo numérico avanzado y manejo de matrices/vectores.
  - np.array(): Convierte una lista u otra estructura en un arreglo matemático de NumPy.
  - np.linspace(): Genera un arreglo de números espaciados uniformemente dentro de un intervalo especificado.
  - np.mean(): Calcula el promedio (media) de los elementos de un arreglo.
  - np.max() / np.min(): Encuentra el valor máximo o mínimo dentro de un arreglo.

ONE-HOT ENCODING
Técnica donde las variables categóricas o de texto se transforman en columnas binarias (0 y 1) para algoritmos matemáticos.

OPERADORES
Símbolos utilizados para realizar cálculos y comparaciones:
  -  + : Suma (o concatenación en strings/listas).
  -  - : Resta.
  -  * : Multiplicación (o repetición en strings).
  -  / : División decimal.
  - // : División entera (devuelve el resultado sin decimales).
  -  % : Módulo (devuelve el residuo de una división).
  - ** : Exponenciación o potencia.
  - == : Operador de comparación de igualdad (verifica si dos valores son idénticos).

PANDAS
Biblioteca de manipulación, limpieza y análisis de datos en tablas.
  - pd.read_csv(): Carga un archivo CSV y lo convierte en un DataFrame.
  - df.head(): Muestra las primeras 5 filas del DataFrame.
  - df.info(): Muestra un resumen del DataFrame (columnas, valores no nulos y tipos de datos).
  - df.describe(): Calcula estadísticas descriptivas básicas (media, percentiles, min, max) de las columnas numéricas.
  - df.drop(): Elimina filas o columnas específicas del DataFrame.
  - df.dropna(): Elimina las filas (o columnas) que contienen valores nulos/vacíos (NaN).
  - df.fillna(): Rellena los valores nulos con un valor especificado.
  - df.astype(): Cambia el tipo de dato de una columna (ej. de string a int).
  - df.rename(): Cambia el nombre de las columnas o índices.
  - df.reset_index(): Reinicia el índice del DataFrame, volviéndolo numérico desde 0.
  - df.value_counts(): Cuenta la frecuencia de valores únicos en una columna.
  - df.unique(): Devuelve una lista con los valores únicos que existen en una columna sin repetir.
  - df.mean() / df.max() / df.min(): Calcula el promedio, el máximo y el mínimo de una serie o columna.

SEABORN
Biblioteca de visualización de datos basada en Matplotlib, con una sintaxis de alto nivel y estilos estéticos.
  - sns.histplot(): Crea un histograma con más opciones visuales que matplotlib.
  - sns.boxplot(): Crea un gráfico de cajas (ideal para visualizar distribuciones y valores atípicos/outliers).
  - sns.scatterplot(): Crea un gráfico de dispersión (puntos) para observar la relación entre dos variables.

SLICING (Segmentación / Rebanado)
Técnica para extraer subsecciones de secuencias (listas, strings, tuplas) usando corchetes [inicio:fin:paso].

STRINGS (str)
Tipo de dato que almacena cadenas de texto.
  - upper(): Convierte todo el texto a MAYÚSCULAS.
  - lower(): Convierte todo el texto a minúsculas.
  - replace(): Reemplaza una subcadena por otra dentro del texto.
  - split(): Divide el texto en una lista de palabras, separadas por espacios o el carácter indicado.
  - find(): Busca una subcadena y devuelve el índice donde comienza (o -1 si no la encuentra).
  - count(): Cuenta cuántas veces aparece un carácter o palabra específica dentro del texto.
  - strip(): Elimina los espacios en blanco (o caracteres invisibles) al principio y al final del texto.

TUPLAS (tuple)
Colección de datos en secuencia similar a la lista, pero inmutable (no se puede modificar luego de crearse). Se declaran con ().

VARIABLES
Espacios de almacenamiento reservados en la memoria a los cuales se les asigna un nombre para guardar y recuperar información.
